```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('A1.png',0)
hist,bins = np.histogram(img.flatten(),256,[0,256])
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
img = cv2.imread('A1.png',0)
equ = cv2.equalizeHist(img)

res = np.hstack((img, equ))

plt.imshow(res,cmap='gray')
#plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
#cv2.imshow('A1.png', res) 
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
img2 = cv2.imread('A2.png',0)
hist,bins = np.histogram(img2.flatten(),256,[0,256])
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
img2 = cv2.imread('A2.png',0)

equ = cv2.equalizeHist(img2)
res = np.hstack((img2, equ))

plt.imshow(res,cmap='gray')
```

As we see in the generated pictures in both the cases, histogram equalisation processes the images to adjust the contrast by modifying the intensity distribution of their histogram leading to a linear CPF(cumulative probability function) trend of the image.
